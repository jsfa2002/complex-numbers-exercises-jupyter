### Exercise 1: Basic Complex Number Operations

```python
import numpy as np
# operaciones basicas de numeros complejos

z1 = 3+4j
z2 = 1-2j
z3 = 5+6j
z4 = 7-8j
z5 = 2-3j
z6 = 1+4j

# Suma en numeros complejos

suma= z1+z2
print("suma:", suma)

# Multiplicacion en numeros complejos
multi= z3*z4
print("Multiplicacion:", multi)

# Division en numeros complejos
div = z5/z6
print("Division:", div)

# Conjugado y módulo de un número complejo
z = 3-4j
conjugado= np.conj(z)
modulo =np.abs(z)
print("Conjugado:", conjugado)
print("Módulo:", modulo)
```

### Exercise 2: Visualizing Complex Operations

```python
import matplotlib.pyplot as plt

# Funcion para graficar nuneros complejos
def plot_complex(z, label):
    plt.plot(z.real, z.imag, 'o', label=label)

# Graficar un numero complejo y su conjugado
z = 3+4j
plot_complex(z, "3 + 4j")
plot_complex(np.conj(z), "Conjugado: 3 - 4j")

plt.xlabel("Parte real")
plt.ylabel("Parte imaginaria")
plt.legend()
plt.grid()
plt.show()

# Extender la funcion para mostrar suma y multiplicaciom
z1 = 2+3j
z2 = 1-2j

# Suma
suma = z1+ z2
plot_complex(z1, "2 + 3j")
plot_complex(z2, "1 - 2j")
plot_complex(suma, "Suma: 3 + 1j")

# Multiplicacion
multiplicacion = z1 *z2
plot_complex(multiplicacion, "Multiplicación: 8 - 1j")

plt.xlabel("Parte real")
plt.ylabel("Parte imaginaria")
plt.legend()
plt.grid()
plt.show()
```

### Exercise 3: Exploring the Mandelbrot Set

```python
import numpy as np
import matplotlib.pyplot as plt

# funcion para calcular el conjunto de Mandelbrot
def mandelbrot(c, max_iter):
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z* z +c
    return max_iter

# Función para generar y graficar el conjunto de Mandelbrot
def plot_mandelbrot(center_x, center_y, zoom_level, max_iter):
    # Definir el rango del plano complejo
    pixels = 800
    x_width = 3/ zoom_level
    y_width = 3 /zoom_level
    x = np.linspace(center_x - x_width/2, center_x + x_width/2, pixels)
    y = np.linspace(center_y - y_width/2, center_y + y_width/2, pixels)
    xx, yy = np.meshgrid(x, y)
    c = xx + yy * 1j

    # este es es para calcular el conjunto de Mandelbrot
    mandelbrot_set = np.array([mandelbrot(ci, max_iter) for ci in c.ravel()]).reshape(c.shape)

    # Graficar
    plt.imshow(mandelbrot_set, extent=[center_x - x_width/2, center_x + x_width/ 2,
                                       center_y - y_width/2, center_y + y_width /2], cmap='hot')
    plt.colorbar()
    plt.title(f"Mandelbrot Set (Zoom: {zoom_level}x, Max Iter: {max_iter})")
    plt.show()

# Ejemplo 1: Conjunto de Mandelbrot normal o estandar
plot_mandelbrot(center_x=-0.5, center_y=0, zoom_level=1, max_iter=30)

# Ejemplo 2: Zoom en una region especcifia
plot_mandelbrot(center_x=-0.7, center_y=0.3, zoom_level=10, max_iter=50)

# Ejemplo 3: Aumentar el numero de iteraciones para mas detalle
plot_mandelbrot(center_x=-0.5, center_y=0, zoom_level=1, max_iter=100)
```
