### Exercise 1: Basic Complex Number Operations

```python
import numpy as np

# 1. Operaciones básicas con números complejos
z1 = 3 + 4j
z2 = 1 - 2j
z3 = 5 + 6j
z4 = 7 - 8j
z5 = 2 - 3j
z6 = 1 + 4j

# Suma
suma = z1 + z2
print("Suma:", suma)

# Multiplicación
multiplicacion = z3 * z4
print("Multiplicación:", multiplicacion)

# División
division = z5 / z6
print("División:", division)

# 2. Conjugado y módulo de un número complejo
z = 3 - 4j
conjugado = np.conj(z)
modulo = np.abs(z)
print("Conjugado:", conjugado)
print("Módulo:", modulo)
```

### Exercise 2: Visualizing Complex Operations

```python
import matplotlib.pyplot as plt

# Función para graficar números complejos
def plot_complex(z, label):
    plt.plot(z.real, z.imag, 'o', label=label)

# Graficar un número complejo y su conjugado
z = 3 + 4j
plot_complex(z, "3 + 4j")
plot_complex(np.conj(z), "Conjugado: 3 - 4j")

plt.xlabel("Parte Real")
plt.ylabel("Parte Imaginaria")
plt.legend()
plt.grid()
plt.show()

# Extender la función para mostrar suma y multiplicación
z1 = 2 + 3j
z2 = 1 - 2j

# Suma
suma = z1 + z2
plot_complex(z1, "2 + 3j")
plot_complex(z2, "1 - 2j")
plot_complex(suma, "Suma: 3 + 1j")

# Multiplicación
multiplicacion = z1 * z2
plot_complex(multiplicacion, "Multiplicación: 8 - 1j")

plt.xlabel("Parte Real")
plt.ylabel("Parte Imaginaria")
plt.legend()
plt.grid()
plt.show()
```

### Exercise 3: Exploring the Mandelbrot Set

```python
import numpy as np
import matplotlib.pyplot as plt

# Función para calcular el conjunto de Mandelbrot
def mandelbrot(c, max_iter):
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z * z + c
    return max_iter

# Función para generar y graficar el conjunto de Mandelbrot
def plot_mandelbrot(center_x, center_y, zoom_level, max_iter):
    # Definir el rango del plano complejo
    pixels = 800
    x_width = 3 / zoom_level
    y_width = 3 / zoom_level
    x = np.linspace(center_x - x_width / 2, center_x + x_width / 2, pixels)
    y = np.linspace(center_y - y_width / 2, center_y + y_width / 2, pixels)
    xx, yy = np.meshgrid(x, y)
    c = xx + yy * 1j

    # Calcular el conjunto de Mandelbrot
    mandelbrot_set = np.array([mandelbrot(ci, max_iter) for ci in c.ravel()]).reshape(c.shape)

    # Graficar
    plt.imshow(mandelbrot_set, extent=[center_x - x_width / 2, center_x + x_width / 2, 
                                       center_y - y_width / 2, center_y + y_width / 2], cmap='hot')
    plt.colorbar()
    plt.title(f"Mandelbrot Set (Zoom: {zoom_level}x, Max Iter: {max_iter})")
    plt.show()

# Ejemplo 1: Conjunto de Mandelbrot estándar
plot_mandelbrot(center_x=-0.5, center_y=0, zoom_level=1, max_iter=30)

# Ejemplo 2: Zoom en una región específica
plot_mandelbrot(center_x=-0.7, center_y=0.3, zoom_level=10, max_iter=50)

# Ejemplo 3: Aumentar el número de iteraciones para más detalle
plot_mandelbrot(center_x=-0.5, center_y=0, zoom_level=1, max_iter=100)
```

### Exercise 4: Creating a Julia Set


```python
# Función para generar el conjunto de Julia
def julia(c, max_iter):
    z = c
    for i in range(max_iter):
        z = z**2 + (-0.4 + 0.6j)
        if abs(z) > 2:
            return i
    return max_iter

# Parámetros
width, height = 800, 800
x_min, x_max = -2, 2
y_min, y_max = -2, 2
max_iter = 100

# Crear la imagen
image = np.zeros((height, width))
for x in range(width):
    for y in range(height):
        real = x_min + (x / width) * (x_max - x_min)
        imag = y_min + (y / height) * (y_max - y_min)
        c = real + imag * 1j
        image[y, x] = julia(c, max_iter)

# Mostrar la imagen
plt.imshow(image, extent=(x_min, x_max, y_min, y_max), cmap='hot')
plt.colorbar()
plt.title("Conjunto de Julia")
plt.show()
```
